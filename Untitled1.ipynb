```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data for CS-01 (Table M5)
distances = [20, 40, 60, 80, 100, 120, 140, 160, 180, 200]
depths = [4.2, 6.8, 7.5, 8.2, 7.8, 6.5, 4.8, 3.2, 2.1, 1.5]
velocities = [0.47, 0.70, 0.81, 0.84, 0.78, 0.64, 0.48, 0.36, 0.29, 0.23]
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
```

```python
# Create a color-mapped scatter plot
sc = ax.scatter(distances, depths, c=velocities, cmap='RdYlGn_r', 
                s=100, edgecolors='black', linewidth=1, vmin=0.2, vmax=1.0)
```

```python
# Add velocity labels
for x, y, v in zip(distances, depths, velocities):
    ax.annotate(f'{v:.2f}', xy=(x, y), xytext=(x+2, y-0.3), fontsize=8, ha='left')
```

```python
# Add isovel contour (simulated)
# Create a coarse grid for contour
X_grid = np.linspace(20, 200, 20)
Z_grid = np.linspace(0, 10, 20)
X_mesh, Z_mesh = np.meshgrid(X_grid, Z_grid)
V_mesh = np.interp(X_grid, distances, velocities)  # simplified interpolation
V_mesh = np.outer(np.ones(20), V_mesh)
contour = ax.contour(X_mesh, -Z_mesh, V_mesh, levels=[0.4, 0.6, 0.8], 
                     colors='black', linewidths=1, linestyles='dashed')
ax.clabel(contour, inline=True, fontsize=9)
```

```python
ax.set_xlabel('Distance from South Bank (m)', fontsize=12)
ax.set_ylabel('Depth (m)', fontsize=12)
ax.set_title('Figure 4.7: Velocity Distribution at Cross-Section CS-01 (250 m from upstream)', fontsize=12)
ax.invert_yaxis()
ax.set_xlim(15, 210)
ax.set_ylim(9, 0)
cbar = plt.colorbar(sc, label='Mean Velocity (m/s)')
```

```python
plt.tight_layout()
plt.savefig('Figure_4.7_CS01_Velocity.png', dpi=300)
plt.show()
```
