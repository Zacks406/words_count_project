```python
import pandas as pd
```

```python
df = pd.read_csv("enable2.txt", header = None, names = ["words"])
```

```python
df
```

```python
df["char_count"] = df["words"].str.len()
```

```python
df
```

```python
df["words"] = df["words"].fillna("").astype(str)
```

```python



def word_value(words):
    return sum(
    ord(char.lower()) - ord("a") + 1
    for char in words
    if char.isalpha()
    )
```

```python
df["value"] = df["words"].apply(word_value)
```
