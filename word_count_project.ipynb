```python
import pandas as pd
```

```python
df = pd.read_csv("enable1.txt", header = None, names = ["words"]);
```

```python
df["word_count"] = df["words"].str.len()
df
```

```python
df.rename(columns = {"word_count":"char_count"}, inplace = True)
```

```python
df
```

```python
df["words"] = df["words"].fillna("").astype(str)
```

```python
def word_value(words):
    return sum(
        ord(char.lower()) - ord("a") + 1
        for char in words
        if char.isalpha()
    )
```

```python
df["value"] = df["words"].apply(word_value)
```
